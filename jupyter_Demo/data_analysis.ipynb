```python
'''
    缺失值的处理 nan：not a number
'''
s=pd.Series([1,2,np.nan,pd.NA,None]) #三种缺失值
df=pd.DataFrame([[1,2,np.nan],[3,pd.NA,None],[4,5,6]],
                columns=['第一列','第二列','第三列'])
# print(s.isna())
# print(s.isnull())
# print(df.isna())
# print(df.isnull())
print('--------------------')
print(df.isna().sum(axis=1)) #第一行一个，第二行两个，第三行没有不显示
print('------------------')
print(s.isna().sum()) #查看缺失值的个数
print('------------------')
print(df.dropna()) #df中剔除行，不是单个的记录一整条
print(df.dropna(how='all')) #如果所有的值是缺失值才会全部删除一整行
print(df.dropna(thresh=2))  #如果至少有n个值不是缺失值就保留
print(df.dropna(subset=['第一列']))  #检测某列的缺失值
```

```python
#填充缺失值
print(df.fillna({'temp_max':20}))  #使用字典填充temp_max是填充的列名，20是具体数字，填充所有的缺失值
print(df.fillna(df[['temp_max','wind']].mean()))  #使用统计值填充
print(df.ffill())   #用前面相邻的填充
print(df.bfill()) #用后面的相邻填充
```

```python
#数据的去重
import numpy as np
import pandas as pd
data={
    'name':['alice','alice','bob','alice','jack','bob'],
    'age':[26,25,30,25,35,30],
    'city':['NY','NY','LA','NY','SF','LA'],
    'gender':['male','female','male','female','male','female']
}
df=pd.DataFrame(data)
df.duplicated() #一整条记录都是一样的，标记为重复
df.drop_duplicates(subset=['name']) #根据指定列去重
df.drop_duplicates(subset=['name'], keep='last')  #留下最后出现的重复数据n
```

```python
# 数据类型的转换

df['age']=df['age'].astype('int16')  #astype类型转换
df['gender']=df['gender'].astype('category')
df['is male']= df['gender'].map({'male':True,'female':False})
df.dtypes
```
